## Lab Exercise: Pathfinding
- Implement the A* algorithm in Python using a 10x10 Grid.
- Create a class 'Node' with 'x, y' coordinates.

```python
import random
import heapq
from dataclasses import dataclass

# ------------------------
# Node Class
# ------------------------
@dataclass(frozen=True)
class Node:
    x: int
    y: int

# ------------------------
# Helper Functions
# ------------------------
def manhattan(a: Node, b: Node):
    return abs(a.x - b.x) + abs(a.y - b.y)

def in_bounds(x, y, size=10):
    return 0 <= x < size and 0 <= y < size

def get_neighbors(node, grid):
    size = len(grid)
    dirs = [(1,0), (-1,0), (0,1), (0,-1)]
    result = []
    for dx, dy in dirs:
        nx, ny = node.x + dx, node.y + dy
        if in_bounds(nx, ny, size) and grid[nx][ny] == 0:
            result.append(Node(nx, ny))
    return result

def print_grid(grid, start=None, goal=None, path=None):
    path = set(path) if path else set()
    for i in range(len(grid)):
        row = ""
        for j in range(len(grid[0])):
            node = Node(i, j)
            if node == start:
                row += " S "
            elif node == goal:
                row += " G "
            elif node in path:
                row += " * "
            elif grid[i][j] == 1:
                row += " # "
            else:
                row += " . "
        print(row)
    print()

# ------------------------
# A* Algorithm
# ------------------------
def a_star(grid, start, goal):
    open_heap = []
    counter = 0

    g = {start: 0}
    f = {start: manhattan(start, goal)}
    came = {start: None}

    heapq.heappush(open_heap, (f[start], counter, start))

    closed = set()
    nodes_visited = 0

    while open_heap:
        _, _, current = heapq.heappop(open_heap)
        nodes_visited += 1

        if current == goal:
            path = []
            while current:
                path.append(current)
                current = came[current]
            return path[::-1], nodes_visited

        if current in closed:
            continue
        closed.add(current)

        for nbr in get_neighbors(current, grid):
            tg = g[current] + 1
            if tg < g.get(nbr, float("inf")):
                came[nbr] = current
                g[nbr] = tg
                f[nbr] = tg + manhattan(nbr, goal)
                counter += 1
                heapq.heappush(open_heap, (f[nbr], counter, nbr))

    return None, nodes_visited

# ------------------------
# Run demo
# ------------------------
size = 10
grid = [[0 if random.random() > 0.25 else 1 for _ in range(size)] for _ in range(size)]

start = Node(0, 0)
goal = Node(9, 9)

grid[start.x][start.y] = 0
grid[goal.x][goal.y] = 0

print("GRID:")
print_grid(grid, start, goal)

path, visited = a_star(grid, start, goal)

print("A* PATH:")
if path:
    print_grid(grid, start, goal, path)
else:
    print("No path found")

print(f"A* visited {visited} nodes")
```
